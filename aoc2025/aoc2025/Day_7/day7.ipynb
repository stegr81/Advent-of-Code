```python
import io
import pandas as pd
```

```python
with open("input_7.txt", "r") as r:
    tx = r.readlines()
    r.close()
```

```python
sep_tx = []
for line in tx:
    line = line.rstrip('\n')
    line = [f'{x} ' for x in line]
    line = (''.join(line)).rstrip(',')
    sep_tx.append(line)
    

data_string = "\n".join(sep_tx)
string_data = io.StringIO(data_string)
df = pd.read_csv(string_data, header=None, delim_whitespace=True)
```

```python
df
```

```python
# part 1
cols_to_check = []
split=0
for col in range(df.shape[1]):
    if df.iat[0,col] == 'S':
        cols_to_check.append(col)
for j in range(df.shape[0]):
    for i in range(df.shape[1]):
        if i in cols_to_check:
            if df.iat[j,i] == '^':
                cols_to_check.append(i-1)
                cols_to_check.append(i+1)
                cols_to_check.remove(i)
                split+=1
                cols_to_check = list(set(cols_to_check))

print(split)
```
